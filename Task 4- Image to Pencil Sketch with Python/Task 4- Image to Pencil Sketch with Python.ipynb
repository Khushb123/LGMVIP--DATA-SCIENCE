## Task 4: Image to Pencil Sketch with Python

### Business case: We need to read the image in RBG format and then convert it to a grayscale image.

```python
import cv2
```

```python
image = cv2.imread("E2zmL5rXMAgV99N.jpg")
cv2.imshow("BTS", image)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.imshow(image, cmap='gray')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
image.max()
```

```python
image.min()
```

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image,cmap='gray')
```

```python
blurr = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
plt.imshow(blurr, cmap='gray')
```

```python
invertedblur = 255-blurr
```

```python
sketch = cv2.divide(gray_image, invertedblur, scale=256.0)
```

```python
cv2.imwrite("outputImage.png", sketch)
```

```python
cv2.imshow("Output Image", sketch)
```

```python
plt.imshow(sketch, cmap='gray')
```
